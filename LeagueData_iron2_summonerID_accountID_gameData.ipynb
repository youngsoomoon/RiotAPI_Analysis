# 추출된 accountId 넣어서 gameId 받아오기

```python
gameId_list = []
for g in accId_list:
    gameId_API = "https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/"+g+'?api_key='+api_key
    r_gameId_API = requests.get(gameId_API)
    gameId_dict = r_gameId_API.json()['matches'][0]
    gameId_list.append(list(gameId_dict.values())[1])
gameId_list
```

# 추출된 gameId 넣어서 FirstBlood 추출하기
## gameId로 gamedata 받아오기
## gamedata중 'teams'의 밸류값만 추출
## 'teams' 밸류값에는 한 게임 안에서 경쟁을 한 두팀의 데이터가 들어있는데 그중 첫번째 팀의 승패 여부와, FirstBlood 여부만 추출

```python
FirstBlood = []
Win = []
for gd in gameId_list:
    gameData_API = "https://kr.api.riotgames.com/lol/match/v4/matches/"+ str(gd) +'?api_key='+api_key
    r_gameData = requests.get(gameData_API)
    gamedata_dict = r_gameData.json()['teams']
    team1 = gamedata_dict[0]
    FirstBlood.append(str(team1['firstBlood']))
    Win.append(str(team1['win']))
print(FirstBlood)
print(Win)
    
```
