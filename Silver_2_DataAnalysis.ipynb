```python
silver2_gameID_list = []
for g in silver2_accID:
    gameId_API = "https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/"+g+'?api_key='+api_key
    r_gameId_API = requests.get(gameId_API)
    silver2_gameId_collection.insert_one(r_gameId_API.json())
    print(r_gameId_API.json())
    time.sleep(1.5)
```

```python
matches_list = []
for i in list(silver2_gameId_collection.find({},{'_id' : 0,'matches':1})):
    matches_list.append(i)
matches_list[0]
```

```python
silver2_gameId = []
for i in matches_list:
    for j in range(len(list(i.values())[0])):
        silver2_gameId.append(list(i.values())[0][j]['gameId'])
silver2_gameId 
```

```python
for i in silver2_gameId:
    gameData_API = "https://kr.api.riotgames.com/lol/match/v4/matches/"+ str(i) +'?api_key='+api_key
    r_gameData = requests.get(gameData_API)
    silver2_gameData_collection.insert_one(r_gameData.json())
    print(silver2_gameId.index(i),'/',len(silver2_gameId))
    time.sleep(1.4)
```
